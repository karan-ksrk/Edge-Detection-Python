```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vertical_filter = [
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
]
```

```python
horizontal_filter = [
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]
```

```python
img = plt.imread('girl.jpg')
n, m, d = img.shape
print(img.shape)
        
```

```python
edges_img = img.copy()

for row in range(3, n-2):
    for col in range(3, m-2):
        local_pixels = img[row-1:row+2, col-1:col+2, 0]
        
        vertical_transform_pixels = vertical_filter*local_pixels
        vertical_score = vertical_transform_pixels.sum()/4
        
        horizontal_transform_pixels = horizontal_filter*local_pixels
        horizontal_score = horizontal_transform_pixels.sum()/4
        
        edge_score = (vertical_score**2 + horizontal_score**2)**.5
        
        edges_img[row, col] = [edge_score]*3
        
edges_img = edges_img/edges_img.max()
```

```python
plt.imshow(img)
```

```python
plt.imshow(edges_img)
```
